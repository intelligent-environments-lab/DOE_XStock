```python
import os
import sys
sys.path.insert(0, '..')
import numpy as np
import pandas as pd
from doe_xstock.data import VersionDatasetType, VersionWeatherData
from doe_xstock.end_use_load_profiles import EndUseLoadProfiles
```

```python
eulp = EndUseLoadProfiles(
    dataset_type=VersionDatasetType.RESSTOCK,
    weather_data=VersionWeatherData.TMY3,
    year_of_publication=2021,
    release=1,
)
```

```python
# eulp.version.clear_cache()
```

```python
metadata = eulp.metadata.metadata.get(filters={
    'in.resstock_county_id': ['GA, Fulton County'],
    'in.geometry_building_type_recs': ['Single-Family Attached'],
})

with pd.option_context('display.max_columns', None):
    display(metadata.head())
    print('Number of represented single-family attached buildings in GA, Fulton County:', metadata.shape[0])
```

```python
BLDG_ID = metadata.sample(1, random_state=0).iloc[0].name
print('Selected building ID:', BLDG_ID)
```

```python
# print(BLDG_ID, 'metadata:')
# display(eulp.get_building(BLDG_ID).metadata)
```

```python
IDD_FILEPATH = '/Applications/EnergyPlus-9-6-0/PreProcess/IDFVersionUpdater/V9-6-0-Energy+.idd'

if os.path.isfile(IDD_FILEPATH):
    print('IDD exists!')
else:
    print('Provided IDD does not exist:', IDD_FILEPATH)
```

```python
# mechanical
SIMULATION_ID = f'simulation-{BLDG_ID}-mechanical'
mechanical_building = eulp.simulate_building(
    BLDG_ID,
    IDD_FILEPATH,
    simulation_id=SIMULATION_ID,
    output_directory=SIMULATION_ID,
    default_output_variables=True,
)
```

```python
# ideal when there is no existing cache but saving to cache afterwards
SIMULATION_ID = f'simulation-{BLDG_ID}-ideal'
ideal_building = eulp.simulate_building(
    BLDG_ID,
    IDD_FILEPATH,
    simulation_id=SIMULATION_ID,
    output_directory=SIMULATION_ID,
    default_output_variables=True,
    ideal_loads=True,
    cache=True,
)
```

```python
# ideal using existing cache
SIMULATION_ID = f'simulation-{BLDG_ID}-ideal'
ideal_building = eulp.simulate_building(
    BLDG_ID,
    IDD_FILEPATH,
    simulation_id=SIMULATION_ID,
    output_directory=SIMULATION_ID,
    default_output_variables=True,
    ideal_loads=True,
    cache=True,
)
```

```python
# ideal using filepath to cached model
SIMULATION_ID = f'simulation-{BLDG_ID}-ideal'
ideal_building = eulp.simulate_building(
    BLDG_ID,
    IDD_FILEPATH,
    simulation_id=SIMULATION_ID,
    output_directory=SIMULATION_ID,
    default_output_variables=True,
    model=ideal_building.simulator.cached_fixed_ideal_loads_idf_filepath,
    osm=False,
    ideal_loads=True,
)
```

```python
# ideal using filepath to cached model
SIMULATION_ID = f'simulation-{BLDG_ID}-ideal'
thermostat_setpoint = pd.DataFrame(np.zeros((8760, 2)))
thermostat_setpoint.loc[:, 0] = 23
thermostat_setpoint.loc[:, 1] = 20
ideal_building = eulp.simulate_building(
    BLDG_ID,
    IDD_FILEPATH,
    simulation_id=SIMULATION_ID,
    output_directory=SIMULATION_ID,
    default_output_variables=True,
    ideal_loads=True,
    cache=True,
    thermostat_setpoint=thermostat_setpoint,
)
```
